## Category Theory
Category theory is the mathematics of abstraction and composition. This can become highly abstract. Thankfully, given the compositional nature of deep learning, categorical constructs have concrete meaning. No prior experience in category theory is needed.

A category is a mathematical space where we have anchors called *objects* and connections called *morphisms*. Morphisms have a *domain* and *codomain* object, akin to functions with sets. A category is equipped with an associative composition operator, denoted $; : \mathcal{C}(X,Y) \times \mathcal{C}(Y,Z) \rightarrow \mathcal{C}(X,Z)$, where $\mathcal{C}(X,Y)$ denotes the collection of morphisms with domain $X$ and codomain $Y$. Compositionality means that $(f;g);h=f;(g;h)$. Each object has an identity morphism, so that $\text{Id}_X: X \rightarrow X$ composes so that $\text{Id}_X ; f = f = f ; \text{Id}_Y$.

Naturally, we implement composition as a construction rule in `../data_structures/ProdCategory`.

## Product Category
We will be dealing with categories equipped with (symmetric monoidal) products. Products allow us to combine objects and morphisms in parallel. Given objects $X$ and $Y$, we form an object called $X \otimes Y$. Given morphisms $f: X \rightarrow Y$ and $g: Z \rightarrow W$, we form $f \otimes g: (X \otimes Z) \rightarrow (Y \otimes W)$.

This structure can be condensed into saying that all objects in $\text{Ob}\mathcal{C}$ are built from the product of lone objects $L$. Therefore, we can use $\otimes: \Pi L \rightarrow \text{Ob}\mathcal{C}$ and $\otimes: \Pi \text{Mo}\mathcal{C} \rightarrow \text{Mo}\mathcal{C}$ as construction rules, `ProdObject[L](content: Prod[L])` and `ProductOfMorphisms[L, M: Morphism[L]](content: Prod[M])`.

The rearrangement of tuples is a particular class of morphisms. As composition is restricted by shapes, we must explicitly indicate when we copy, swap, project, or delete out of a tuple. This is done by a `Rearrangement[L](mapping: Prod[int], _dom: Prod[L])` class. The $j^{th}$ output is given by the $\texttt{mapping}[j]$ input. This corresponds to a mathematical construct that takes a map $\mu: J \rightarrow I$ and a domain $\Pi_{i \in I} X_i$ so that $[\mu]: \Pi_{i \in I} X_i \rightarrow \Pi_{j \in J} X_{\mu(j)}$ obeys $(\mathbf{x})[\mu][j]=\mathbf{x}[\mu(j)]$.

## Encoding a Category
A product category is built from sequential composition and parallel products. We also allow for a wrapper `Block`, though this is mainly for organizational / aesthetic reasons and to enable for-loops. These constructs build on **seed morphisms**, which vary from category to category. Specific seed morphisms are not defined in `../data_structures/ProdCategory`. Once a seed morphism type is plugged into a category, composition and parallel products flow naturally. A category can thus be seen as the recursive type;

```python
type ProdCategory[L, M: Morphism[L]] = (
      Composed[L, ProdCategory[L, M]]
    | Block[L, ProdCategory[L, M]]
    | ProductOfMorphisms[L, ProdCategory[L, M]]
    | Rearrangement[L]
    | M )
```

## Diagramming a Category
Just as we can implement mathematical construction rules in Python, we can implement them as diagrams. The same principles apply: a diagrammatic construction rules references the constructors in a clear manner, references metadata, or references a pending variable. We diagram objects as wires, and morphisms as blocks with their domain to the left and codomain to the right. Identities then can be diagrammed as uninterrupted wires. We associate sequential composition with horizontal placement, and parallel products with vertical placement. This naturally enforces the rule that the product of morphisms yields a morphism with the product of their dom/cod.
